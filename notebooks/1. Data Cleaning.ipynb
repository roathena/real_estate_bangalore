```python
df2['bhk'] = df2['size'].apply(lambda x: x.split(' ')[0])
```

```python
df3 = df2.drop(['size'], axis=1)
```

```python
df3['total_sqft'].unique()
```

```python
def is_float(x):
    try: 
        float(x)
    except Exception as e:
        return False
    return True
```

```python
df3[~df3.total_sqft.apply(is_float)].head(10)
```

```python
def avg_sqft(x):
    tokens = x.split(' - ')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4.total_sqft = df4.total_sqft.apply(lambda x: avg_sqft(x))
```

```python
df4.to_csv('../data/final.csv', index=False)
```
